```python
from scipy.optimize import minimize_scalar
from scipy.stats import gamma
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X = np.arange(10)
    
def periodic(x, A=1, B=1, C=0, D=0):
    return A * np.sin(B*x - C) + D

def decaying(x, w, b, target):
    return np.exp(w * x + b) + target

def spike(n_samples, theta, b):
    return np.where(np.random.random(n_samples) < theta, b, 0)

spiker = spike(X.shape[0], 0.005, 0.75)
decayer = decaying(X, -0.01, -0.3, 0.1)
perioder = periodic(X, 0.5, 1, 0, 0.5)
#plt.plot(X*5, perioder)
#plt.plot(X*5, decayer)
plt.plot(X, decayer * (0.5 + perioder))
```

```python
class Schedule:
    def __init__(self, start_value):
        self.t = 0
        self.value = start_value

    def step(self):
        self.t += 1

    def get(self):
        return max([0, min([1., self.value])])


class ScopingPeriodic(Schedule):
    def __init__(self, start_value=0.7, target_value=0.1, decay_factor=-0.01,
                 period=0.25, restart_prob=0.001):
        super().__init__(start_value)
        self.start_value = start_value
        self.target_value = target_value
        self.decay_factor = decay_factor
        self.period = period
        self.restart_prob = restart_prob

    def _decay(self, t):
        return np.exp(self.decay_factor * t + (self.start_value - 1)) + self.target_value

    def _periodic(self, t):
        return 0.5 * np.sin(self.period * t) + 0.5

    def step(self):
        if np.random.random() <= self.restart_prob:
            self.t = 1
        self.value = self._decay(self.t) * (0.5 * self._periodic(self.t))
        self.t += 1
```

```python
schedule = ScopingPeriodic()
```

```python
epsilons = []
for i in range(10000):
    epsilons.append(schedule.get())
    schedule.step()
```

```python
n = 500
plt.plot(np.arange(n), epsilons[:n])
```
